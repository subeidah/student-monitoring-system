# mock test

```python
import sqlite3 #importing library
```

```python
def change_datatype(df): #function to change datatype
    """Changes datatype of specified columns in any dataframe"""
    df['Started_on']= pd.to_datetime(df['Started_on'])
    df['Completed']= pd.to_datetime(df['Completed']) 
```

```python
change_datatype(dfFormattedCleanMocktest) #using function to update datatype for all tests
change_datatype(dfFormattedCleanTest_1)
change_datatype(dfFormattedCleanTest_2)
change_datatype(dfFormattedCleanTest_3)
change_datatype(dfFormattedCleanTest_4)
change_datatype(dfFormattedCleanSumTest)
```

```python
name_db = 'ResultsDatabase' # creating global varibale to use in function 
```

```python
def save_students_marks_into_db(df, name_db, name): 
    """Saves dataframes to database"""
    sqliteConnection = sqlite3.connect('ResultDatabase.db')
    
    df.to_sql(name, 
             sqliteConnection,
             if_exists = 'replace',
             index=False)

    sqliteConnection.close()

save_students_marks_into_db(dfFormattedCleanMocktest, name_db, 'MockTest') #using function to save dfs to db
save_students_marks_into_db(dfFormattedCleanTest_1, name_db, 'FormativeTest_1')
save_students_marks_into_db(dfFormattedCleanTest_2, name_db, 'FormativeTest_2')
save_students_marks_into_db(dfFormattedCleanTest_3, name_db, 'FormativeTest_3')
save_students_marks_into_db(dfFormattedCleanTest_4, name_db, 'FormativeTest_4')
save_students_marks_into_db(dfFormattedCleanSumTest, name_db,'SumTest')
```

# 1.5 Modify Data

```python
import numpy as np #importing library
```

```python
dfFormattedCleanSumTest[0:60]
```

```python
chosen_idx = np.random.choice(150,20, replace = False) #randomly generating 20 numbers between 0-150

chosen_idx
```

```python
import random
x = random.sample(range(4,16),3) #randomly generating 3 numbers to use for columns

x 
```

```python
a = np.random.uniform(0,1) #generating random float between 0 and 1 to times results by 
b = np.random.uniform(0,1)
c = np.random.uniform(0,1)

print(a)
print(b)
print(c)
```

```python
#changing the random columns generated by * by the random float created for each question column.
#This is done for the 20 selected rows
dfFormattedCleanSumTest.iloc[chosen_idx,x[0]] = a * dfFormattedCleanSumTest.iloc[chosen_idx,x[0]] 
dfFormattedCleanSumTest.iloc[chosen_idx,x[1]] = b * dfFormattedCleanSumTest.iloc[chosen_idx,x[1]]
dfFormattedCleanSumTest.iloc[chosen_idx,x[2]] = c * dfFormattedCleanSumTest.iloc[chosen_idx,x[2]]
```

```python
#updating the grade column, with different weighted questions.
dfFormattedCleanSumTest["Grade"] = dfFormattedCleanSumTest["Q1"] * 0.5 + dfFormattedCleanSumTest["Q2"] * 0.3 + dfFormattedCleanSumTest["Q3"] * 0.6 + dfFormattedCleanSumTest["Q4"] *0.7 + dfFormattedCleanSumTest["Q5"] *0.4 + dfFormattedCleanSumTest["Q6"] * 0.5+ dfFormattedCleanSumTest["Q7"] * 1.5+ dfFormattedCleanSumTest["Q8"] * 1.5+ dfFormattedCleanSumTest["Q9"] * 1.5+ dfFormattedCleanSumTest["Q10"] * 1.0 + dfFormattedCleanSumTest["Q11"] *0.4 + dfFormattedCleanSumTest["Q12"]* 0.5 + dfFormattedCleanSumTest["Q13"] * 0.6
dfFormattedCleanSumTest["Grade"] = dfFormattedCleanSumTest["Grade"] / 10


dfFormattedCleanSumTest[0:60]
```
